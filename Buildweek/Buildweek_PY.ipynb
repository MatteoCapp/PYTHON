```python
import numpy as np

# List of the 10 largest Italian cities
italian_cities = [
    "Rome", "Milan", "Naples", "Turin", "Palermo",
    "Genoa", "Bologna", "Florence", "Bari", "Catania"
]

# Function to generate a selling price
def generate_selling_price(base_price):
    base_price = float(base_price.replace('$', '').strip())
    selling_price = np.random.uniform(base_price + 10, 300)
    # Round to the nearest 0.99
    selling_price = round(selling_price / 10) * 10 - 0.01
    return f"${selling_price:.2f}"

# Function to generate a number of items sold
def generate_items_sold():
    return np.random.randint(1, 101)

# Function to generate a random city
def generate_city():
    return np.random.choice(italian_cities)

# Create two dataframes for AWLab and Foot Locker
df_awlab = scarpe_indice.copy()
df_footlocker = scarpe_indice.copy()

# Add new columns to the dataframes
for dataframe in [df_awlab, df_footlocker]:
    dataframe['Number of Items Sold'] = [generate_items_sold() for _ in range(len(dataframe))]
    dataframe['Selling Price'] = [generate_selling_price(price) for price in dataframe['Price (USD)']]
    dataframe['City'] = [generate_city() for _ in range(len(dataframe))]

# Save the dataframes to CSV files
awlab_file_path = '/Users/matte/Desktop/Data analyst corso/PYTHON/Buildweek/AWLab_sales.csv'
footlocker_file_path = '/Users/matte/Desktop/Data analyst corso/PYTHON/Buildweek/FootLocker_sales.csv'

df_awlab.to_csv(awlab_file_path, index=False)
df_footlocker.to_csv(footlocker_file_path, index=False)

awlab_file_path, footlocker_file_path
```

```python
#Calcolo del numero totale di articoli venduti per materiale in AWLab
aw_material_sales = aw.groupby('Material')['Number of Items Sold'].sum().reset_index()
aw_material_sales.rename(columns={'Number of Items Sold': 'Total_Items_Sold'}, inplace=True)

#Calcolo delle percentuali di materiali venduti
aw_material_sales['Percentage_Sold'] = (aw_material_sales['Total_Items_Sold'] / aw_material_sales['Total_Items_Sold'].sum()) * 100

#Ordinare i dati per numero totale di articoli venduti
aw_material_sales = aw_material_sales.sort_values(by='Total_Items_Sold', ascending=True)

#Grafico per AWLab
fig, ax = plt.subplots(figsize=(12, 8))

#Creare un grafico a barre orizzontali
bars = ax.barh(aw_material_sales['Material'], aw_material_sales['Total_Items_Sold'], color='blue', alpha=0.7)

#Annotare le percentuali sulle barre
for bar, perc in zip(bars, aw_material_sales['Percentage_Sold']):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{perc:.2f}%', va='center')

ax.set_xlabel('Numero Totale di Articoli Venduti')
ax.set_title('AWLab - Numero Totale e Percentuale di Articoli Venduti per Materiale')

plt.show()
```

```python
#Calcolo del numero totale di articoli venduti per materiale in FootLocker
ft_material_sales = ft.groupby('Material')['Number of Items Sold'].sum().reset_index()
ft_material_sales.rename(columns={'Number of Items Sold': 'Total_Items_Sold'}, inplace=True)

#Calcolo delle percentuali di materiali venduti
ft_material_sales['Percentage_Sold'] = (ft_material_sales['Total_Items_Sold'] / ft_material_sales['Total_Items_Sold'].sum()) * 100

#Ordinare i dati per numero totale di articoli venduti
ft_material_sales = ft_material_sales.sort_values(by='Total_Items_Sold', ascending=True)

#Grafico per FootLocker
fig, ax = plt.subplots(figsize=(12, 8))

#Creare un grafico a barre orizzontali
bars = ax.barh(ft_material_sales['Material'], ft_material_sales['Total_Items_Sold'], color='green', alpha=0.7)

#Annotare le percentuali sulle barre
for bar, perc in zip(bars, ft_material_sales['Percentage_Sold']):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{perc:.2f}%', va='center')

ax.set_xlabel('Numero Totale di Articoli Venduti')
ax.set_title('FootLocker - Numero Totale e Percentuale di Articoli Venduti per Materiale')

plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from geopy.geocoders import Nominatim
import time
```

```python
# Verifica che la colonna 'fatturato' sia presente
print("Colonne presenti nel DataFrame:", aw.columns)

# Funzione per ottenere le coordinate delle città
def get_coordinates(city):
    geolocator = Nominatim(user_agent="my_unique_application")
    location = geolocator.geocode(city + ", Italy")
    if location:
        return location.latitude, location.longitude
    else:
        return None, None

# Aggiungi colonne di latitudine e longitudine al DataFrame
coordinates = []
for city in aw['City']:
    coordinates.append(get_coordinates(city))
    time.sleep(1)  # Aggiungi un ritardo di 1 secondo tra le richieste

aw[['Latitude', 'Longitude']] = pd.DataFrame(coordinates, columns=['Latitude', 'Longitude'])

# Filtra i dati con latitudine e longitudine non nulli
aw = aw.dropna(subset=['Latitude', 'Longitude'])

# Verifica che il DataFrame aw contenga ancora tutte le colonne
print("Colonne presenti nel DataFrame dopo aver aggiunto coordinate:", aw.columns)

# Carica la mappa del mondo da un file SHP scaricato
world = gpd.read_file('/Users/matte/Desktop/Data analyst corso/PYTHON/Buildweek/geolocaliz/ne_110m_admin_0_countries.shp')

# Filtra per ottenere solo l'Italia
italy = world[world['NAME'] == "Italy"]

# Crea una GeoDataFrame con i punti di vendita
gdf = gpd.GeoDataFrame(aw, geometry=gpd.points_from_xy(aw.Longitude, aw.Latitude))

# Verifica che il GeoDataFrame contenga tutte le colonne necessarie
print("Colonne presenti nel GeoDataFrame:", gdf.columns)

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(10, 15))
italy.plot(ax=ax, color='white', edgecolor='black')

# Aggiungi i punti delle città
gdf.plot(ax=ax, color='red', markersize=gdf['fatturato'] / gdf['fatturato'].max() * 1000, alpha=0.6)

# Aggiungi titoli e etichette
plt.title('Fatturato per città in Italia')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.show()
```
