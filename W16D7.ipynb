```python
#dal database  stockdata.csv
#Visualizziamo l'andamento delle azioni della Apple ("AAPL")
#Visualizziamo l'andamento di tutte le azioni del dataset
#Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM") negli anni 2012 e 2013
#Con un istogramma, troviamo il range di valori che le azioni di Microsoft hanno assunto pi√π spesso
stockdata = pd.read_csv('/Users/matte/Desktop/Data analyst corso/PYTHON/database vari/stockdata.csv')
stockdata.sample(5)
```

```python
#Visualizziamo l'andamento delle azioni della Apple ("AAPL")
stockdata['Date'] = pd.to_datetime(stockdata['Date'])
plt.figure(figsize=(12, 9))
stockdata['Year'] = stockdata['Date'].dt.year
sns.lineplot(x='Year', y='AAPL', data=stockdata)
plt.show()
```

```python
#Visualizziamo l'andamento di tutte le azioni del dataset
stockdata.index = pd.to_datetime(stockdata.index)
plt.figure(figsize=(12, 9))
sns.lineplot(x='Year', y='GSPC', data=stockdata, label='GSPC')
sns.lineplot(x='Year', y='AAPL', data=stockdata, label='AAPL')
sns.lineplot(x='Year', y='MSFT', data=stockdata, label='MSFT')
sns.lineplot(x='Year', y='IBM', data=stockdata, label='IBM')
sns.lineplot(x='Year', y='SBUX', data=stockdata, label='SBUX')

plt.show()
```

```python
#Visualizziamo l'andamento di Microsoft ("MSFT") e IBM ("IBM") negli anni 2012 e 2013
stockdata['Year'] = stockdata.index.year
anni = stockdata[(stockdata['Year'] == 2012) | (stockdata['Year'] == 2013)]
sns.lineplot(x='Year', y='IBM', data=anni, label='IBM')
sns.lineplot(x='Year', y='MSFT', data=anni, label='MSFT')
plt.title('Andamento MSFT e IBM negli anni 2012 e 2013')
plt.xlabel('Anno')
plt.ylabel('Valore Azioni')
plt.legend(title='Legenda')
plt.show()
```

```python
#Con un istogramma, troviamo il range di valori che le azioni di Microsoft hanno assunto pi√π spesso
sns.histplot(stockdata['MSFT'], bins=30, kde=True)
plt.show()
```

```python
import os
import dotenv
import pandas as pd
from sqlalchemy import create_engine

dotenv.load_dotenv(override=True, dotenv_path=".env")

username= os.getenv("username")
password= os.getenv("password")
host= os.getenv("host")
dbname= "Titanic"

conn_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
db_engine = create_engine(conn_string)
query= "SELECT * FROM titanic"
titanic= pd.read_sql(sql=query, con=db_engine)
titanic.sample(5)
```

```python
#Quanti ponti c'erano sulla nave?
#Visualizzare un grafico con il numero di passeggeri di ogni classe di imbarco

numero_passeggeri = titanic.groupby('PClass')['PassengerId'].count()
classe_imbarco = titanic['PClass'].unique()
plt.bar(classe_imbarco, numero_passeggeri)

plt.ylabel('Numero di passeggeri')
plt.xlabel('Classe di imbarco')
plt.title('numero di passeggeri di ogni classe di imbarco')
plt.show()
```

```python
#Fare la stessa cosa per la colonna alive
numero_passeggeri = titanic.groupby('Survived')['PassengerId'].count()
sopravvissuti = titanic['Survived'].unique()
plt.bar(sopravvissuti, numero_passeggeri)

plt.ylabel('Numero di passeggeri')
plt.xlabel('sopravvissuti')
plt.title('numero di passeggeri ')
plt.show()
```

```python
#Qual era la distribuzione delle tariffe (fare)?
plt.figure(figsize=(15, 10))
numero = titanic.groupby('Fare')['Fare'].count()
plt.hist(titanic['Fare'], bins=100, edgecolor='black')

plt.title('Distribuzione delle tariffe sul Titanic')
plt.xlabel('Tariffe')
plt.ylabel('Frequenza')
ticks_to_show = list(range(int(titanic['Fare'].min()), int(titanic['Fare'].max()) + 1, 10))
plt.xticks(ticks_to_show, rotation=45)
plt.show()
```

```python
#Riusciamo a vedere la distribuzione delle et√† dei passeggeri rispetto alla classe di imbarco? Proviamo con un boxplot e con un violinplot
plt.figure(figsize=(10, 6))

plt.boxplot([titanic[titanic['PClass'] == 1]['Age'],
             titanic[titanic['PClass'] == 2]['Age'],
             titanic[titanic['PClass'] == 3]['Age']],
            labels=[1, 2, 3], patch_artist=True)

plt.title('Confronto della distribuzione delle et√† dei passeggeri per classe di imbarco')
plt.xlabel('Classe di imbarco')
plt.ylabel('Et√†')
plt.show()
```

```python
sns.violinplot(x='PClass', y='Age', data=titanic)
plt.title('Distribuzione delle et√† dei passeggeri rispetto alla classe di imbarco')
plt.xlabel('Classe di imbarco')
plt.ylabel('Et√†')
plt.show()
```

```python
#Visualizziamo un boxplot rispetto alle colonne fare e survived; che cose ne deduciamo?
plt.figure(figsize=(10, 6))

plt.boxplot([titanic[titanic['Survived'] == 0]['Fare'],
             titanic[titanic['Survived'] == 1]['Fare']],
            labels=['sopravvisuto', 'deceduto'], patch_artist=True)

plt.title('Confronto della distribuzione delle et√† dei passeggeri per classe di imbarco')
plt.xlabel('stato')
plt.ylabel('tariffa')
plt.show()
```
