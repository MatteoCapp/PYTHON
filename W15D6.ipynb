```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd

username = str(os.getenv("username"))
password = str(os.getenv("password"))
host = str(os.getenv("host"))
dbname = str(os.getenv("dbname"))

dotenv.load_dotenv(override=True)
conn_string = "mysql+pymysql://" + username + ":" + password + "@" + host + "/" + dbname
db_engine = create_engine(conn_string)

query = "SELECT * FROM dimproduct"
df = pd.read_sql(sql=query, con=db_engine)
```

```python
df.sample(5)
```

```python
print (round(df.DealerPrice,2))
print (round(df.DealerPrice,0))
```

```python
#Utilizzando il metodo .clip(), facciamo in modo che i valori siano compresi tra un minimo di 0 e un massimo di 1000
clipped_df = pd.DataFrame()
for col in df.columns:
    if pd.api.types.is_numeric_dtype(df[col]):  # Se la colonna contiene numeri
        clipped_df[col] = df[col].clip(lower=0, upper=1000)  # Applica il clip
    else:  # Per tutte le altre colonne (stringhe, datetime, ecc.)
        clipped_df[col] = df[col]  # Copia i valori così come sono

# Visualizziamo il DataFrame risultante
print(clipped_df)
```

```python
#Nella directory dei beginner_datasets, quali sono i dataset che contengono dati nulli?
import os
directory_path = "/Users/matte/Desktop/Data analyst corso/PYTHON/beginner_datasets/"
files_with_nulls = []
files = os.listdir(directory_path)

for file in files:
    file_path = os.path.join(directory_path, file)
    if file.endswith('.csv'):
        try:
            df = pd.read_csv(file_path)
            if df.isnull().values.any():
                files_with_nulls.append(file)
        except Exception as e:
            print(f"Errore nel caricamento di {file}: {e}")
print("File che contengono dati nulli:")
for file_with_nulls in files_with_nulls:
    print(file_with_nulls)
```
